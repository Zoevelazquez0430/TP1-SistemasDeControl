Repasito:
Los márgenes de estabilidad miden qué tan lejos está el sistema de volverse inestable. Los calculamos sobre la función de transferencia a lazo abierto G(s).

Margen de Fase (MF): cuántos grados de fase adicional podemos agregar antes de llegar a la inestabilidad. Se mide en la frecuencia de cruce de ganancia ωcg (donde |G(jω)| = 1).
Margen de Ganancia (MG): cuánto podés aumentar la ganancia antes de inestabilizar. Se mede en la frecuencia de cruce de fase ωcp (donde ∠G(jω) = −180°).

Para que el sistema sea estable (fase mínima): MF > 0° y MG > 0 dB.

### Cálculo analítico

La función de transferencia obtenida es:

$$
G(s) = \frac{K_t}{s\bigl(LJ s^2 + (JR + BL)s + (BR + K_e K_t)\bigr)}
$$

Si sustituímos los valores numéricos del enunciado:

$$
G(s) = \frac{0.25}{s\left(0.003\,s^2 + 0.0515\,s + 0.200\right)}
$$

---

### Análisis del sistema

- El sistema es de **tipo 1**, ya que tiene un polo en el origen ($s = 0$).
- Esto implica que la **fase inicial** es de aproximadamente $-90^\circ$.
- La función tiene **3 polos en total**, por lo que:

$$
\lim_{\omega \to \infty} \angle G(j\omega) = -270^\circ
$$

---

### Conclusión

Como la fase comienza en $-90^\circ$ y tiende a $-270^\circ$, necesariamente:

- Existe una frecuencia donde la fase alcanza $-180^\circ$
- Por lo tanto, el sistema tiene un **margen de ganancia finito**

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# Parámetros del enunciado
L = 0.15; R = 0.7; K_e = 0.1
J = 0.02; B = 0.25; K_t = 0.25

# Función de transferencia que calculó tici
G = ctrl.tf([K_t], [L*J, J*R + B*L, B*R + K_e*K_t, 0])

# Márgenes
gm, pm, wcp, wcg = ctrl.margin(G)
gm_dB = 20 * np.log10(gm)

print(f"Margen de Ganancia: {gm_dB:.2f} dB  (en ωcp = {wcp:.2f} rad/s)")
print(f"Margen de Fase:     {pm:.2f}°  (en ωcg = {wcg:.2f} rad/s)")

ctrl.bode(G, dB=True)
plt.suptitle("Bode de G(s) a lazo abierto")
plt.show()
```

### Conclusión sobre la necesidad de compensador

Los márgenes obtenidos son:

- **Margen de Fase:** 72.22° (en ωcg = 1.22 rad/s)
- **Margen de Ganancia:** 22.76 dB (en ωcp = 8.16 rad/s)

Ambos superan los valores recomendados (MF entre 30°–60°, MG > 6 dB).
El sistema es estable en lazo cerrado sin compensación.

Sin embargo, el MF elevado indica un sistema **muy amortiguado y posiblemente lento**,
lo que puede dificultar cumplir con las especificaciones de tiempo de establecimiento
requeridas en la Parte B. Esto motivará el diseño de un compensador más adelante.

Inciso C - Zoe

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
L = 0.15
R = 0.7
K_e = 0.1
J_val = 0.02
B_val = 0.25
K_t = 0.25

# Planta completa G(s)
G = ctrl.tf([K_t], [L*J_val, J_val*R + B_val*L, B_val*R + K_e*K_t, 0])

# Parte mecánica: 1 / s(Js + B)
G_mec = ctrl.tf([1], [J_val, B_val, 0])

# Lazo cerrado ante referencia
G_cl = ctrl.feedback(G, 1)

# Transferencia perturbación → salida
# G_pert = G_mec / (1 + G)
# Lo calculamos directamente como fracción de polinomios
num_pert = G_mec.num[0][0]
den_pert = (G_mec * (1 + G)).den[0][0]  # G_mec * (1 + G) en denominador
G_pert = ctrl.tf(num_pert, ctrl.tf(num_pert, [1]).den[0][0])  # placeholder

# Forma más segura: usar minreal y operar directo
G_pert = ctrl.minreal(G_mec / (1 + G))

t = np.linspace(0, 30, 3000)

_, y_ref  = ctrl.step_response(G_cl, T=t)
_, y_pert = ctrl.step_response(0.1 * G_pert, T=t)

y_total = y_ref + y_pert

plt.figure(figsize=(10,5))
plt.plot(t, np.ones_like(t), 'k--', label='Referencia')
plt.plot(t, y_ref, 'g--', alpha=0.6, label='Solo referencia')
plt.plot(t, y_total, 'b', label='Salida total (ref + perturbación 0.1)')
plt.xlabel('Tiempo [s]')
plt.ylabel('θ(t) [rad]')
plt.title('Respuesta con escalón unitario + perturbación constante de 0.1')
plt.legend()
plt.grid(True)
plt.show()

# Verificación analítica del error
print(f"Error por perturbación en régimen permanente: {y_pert[-1]:.6f} rad")
```

Lo que simulamos fue la superposición de dos efectos:

Respuesta a la referencia (escalón unitario)
Respuesta a la perturbación (escalón de 0.1)

Y el resultado fue que la salida converge a 1.32 rad, no a 1.0. Hay un error permanente de 0.32 rad.
¿Por qué? Porque el controlador es C(s) = 1 (proporcional puro, sin integral). El torque externo constante desplaza la posición, y sin acción integral el sistema no tiene "memoria" para corregirlo completamente.

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
L = 0.15
R = 0.7
K_e = 0.1
J_val = 0.02
B_val = 0.25
K_t = 0.25

# Planta y parte mecánica
G = ctrl.tf([K_t], [L*J_val, J_val*R + B_val*L, B_val*R + K_e*K_t, 0])
G_mec = ctrl.tf([1], [J_val, B_val, 0])

# Lazo cerrado ante referencia
G_cl = ctrl.feedback(G, 1)

# Transferencia perturbación → salida
G_pert = ctrl.minreal(G_mec / (1 + G))

t = np.linspace(0, 30, 3000)

# Rampa unitaria como señal de entrada
rampa = t  # r(t) = t

# Respuesta ante rampa: usamos forced_response
t_r, y_r = ctrl.forced_response(G_cl, T=t, U=rampa)

# Respuesta ante perturbación escalón 0.1
_, y_pert = ctrl.step_response(0.1 * G_pert, T=t)

# Salida total
y_total = y_r + y_pert

plt.figure(figsize=(10, 5))
plt.plot(t, rampa, 'k--', label='Referencia (rampa unitaria)')
plt.plot(t, y_r, 'g--', alpha=0.6, label='Solo referencia')
plt.plot(t, y_total, 'b', label='Salida total (rampa + perturbación 0.1)')
plt.xlabel('Tiempo [s]')
plt.ylabel('θ(t) [rad]')
plt.title('Respuesta con rampa unitaria + perturbación constante de 0.1')
plt.legend()
plt.grid(True)
plt.show()

# Error en régimen permanente
error_rampa = rampa - y_r
print(f"Error por rampa en régimen permanente: {error_rampa[-1]:.4f} rad")
print(f"Error por perturbación en régimen permanente: {y_pert[-1]:.4f} rad")
```

A diferencia de una referencia escalón (donde se desea alcanzar un valor fijo y mantenerse),
una referencia tipo rampa representa una posición angular que **crece constantemente en el tiempo**:

$$r(t) = t \quad \Rightarrow \quad R(s) = \frac{1}{s^2}$$

Esto modela situaciones donde la hélice debe seguir un objetivo en movimiento a velocidad constante.

### Resultado de la simulación

Se simuló el sistema en lazo cerrado con realimentación unitaria y C(s) = 1, sometido
simultáneamente a una referencia tipo rampa unitaria y una perturbación de carga constante de 0.1.

Los errores en régimen permanente obtenidos fueron:

- **Error por rampa:** 0.80 rad
- **Error por perturbación:** 0.32 rad

#### Error ante la rampa

El sistema es de **tipo 1** — tiene un único integrador en la función de transferencia,
proveniente de la relación cinemática $\omega = d\theta/dt$.

Para seguir una rampa con error nulo se necesitaría un sistema de **tipo 2** (dos integradores).
Con un solo integrador, el sistema puede seguir la rampa pero siempre con un desfasaje constante:
la salida crece a la misma velocidad que la referencia, pero nunca la alcanza.

Esto se observa claramente en el gráfico: las curvas son paralelas a la rampa de referencia,
pero desplazadas hacia abajo en 0.80 rad.

#### Error ante la perturbación

Como se analizó en el inciso c, la perturbación de carga constante produce un error permanente
de 0.32 rad que se suma al error por rampa. Esto ocurre porque la perturbación ingresa
después del integrador de posición, y el controlador proporcional puro C(s) = 1
no tiene capacidad de corregirla completamente.

---

### Resumen del comportamiento sin compensación

| Entrada | Error en régimen permanente |
|--------|----------------------------|
| Escalón unitario | 0 rad ✓ |
| Escalón unitario + perturbación 0.1 | 0.32 rad ✗ |
| Rampa unitaria | 0.80 rad ✗ |
| Rampa unitaria + perturbación 0.1 | ~1.12 rad ✗ |

El sistema con C(s) = 1 resulta **insuficiente** para cumplir con las especificaciones
de seguimiento y rechazo de perturbaciones requeridas. Esto motiva el diseño de un
controlador adecuado en el inciso e.
